# **Pregunta - 2:**

Proceso de limpieza. Aplica el proceso de limpieza que consideres adecuado.











*  El corpus de  `WordNet` nos proporciona sinónimos, antónimos, hipónimos, hiperónimos y más.
* Las `stopwords` son palabras comunes que a menudo se filtran del texto durante las tareas de procesamiento del lenguaje natural. Incluyen palabras como `y`, `el`, `es`, etc. `NLTK` proporciona palabras vacías para varios idiomas.

```
# Descarga el corpus de WordNet de NLTK.
nltk.download('wordnet')

# Descargamos las "stopwords" en varios idiomas.
nltk.download('stopwords')

# eDescargamos un tokenizador para dividr el texto en enunciados mediante un modelo no-supervisado.
nltk.download('punkt')
```

## Stopwords

```
# # Create a list `stop_words` that contains the English stop words list
stopwords = stopwords.words('english')

# Se utiliza listado de palabras que indican negación en inglés de ejercicio semana pasada para definir las stopwords.
negwords = [ 'no', 'nor', 'not', 'ain', 'aren', "aren't", 'don', "don't", 'couldn', "couldn't", 'didn', "didn't", 'doesn', "doesn't", 'hadn', "hadn't", 'hasn', "hasn't", 'haven', "haven't", 'isn', "isn't", 'mightn', "mightn't", 'mustn', "mustn't", 'needn', "needn't", 'shan', "shan't", 'shouldn', "shouldn't", 'wasn', "wasn't", 'weren', "weren't", 'won', "won't", 'wouldn', "wouldn't"]

# Se agregan stopwords adicionales que no estaban incluidas en stopwords de nltk, para personalizar nuestra lista de stopwprds.
extra_stopwords = ['us', 'since', 'even', 'lot', 'everywhere', 'elsewhere', 'anywhere', 'somewhere', 'every', 'whoever', 'whenever', 'wherever', 'whatever', 'ever','someway', 'somehow', 'anyway', 'everyway', 'would', 'yet']

# Se eliminan palabras numéricas y colores comunes que no aportan información al Análisis de sentimiento de los comentarios.
number_words = ["one", "two", "three", "four", "five", "six", "seven", "eight", "nine", "ten","black", "white", "red", "blue", "green", "yellow", "purple", "pink", "orange", "gray"]

# Usamos set para no repetir palabras
stopwords = list(set(stopwords) | set(negwords) | set(extra_stopwords) | set(number_words))

# Inspeccionamos las stopwords
print('Numero de stopwords: ', len(stopwords))
print('Lista de stopwords: ', sorted(stopwords))
```

## Tokenization

```
def clean_tok(doc):

  tokens = [] # Array temporal para guardar tokens mientras se limpian.

  for docs in doc.split():
    lower = docs.lower()                       # Minúsculas: Simplificar todo a minúsculas.
    alpha = re.sub('[^a-z]', '', lower)               # Caracteres alfabéticos: Solo considerar caracteres alfabéticos.
    lenght = alpha if len(alpha.strip()) > 1 else ''     # Longitud mayor a 1: Se deben considerar solamente tokens de longitud mayor a 1.
    word = lenght if lenght not in stopwords else '' # Stopwords: Eliminar las stopwords.
    if word:
      tokens.append(word)                       # Tokenización: Cada comentario queda tokenizado por palabras, o cadena de caracteres alfabéticos.

  return tokens
```

## Stemming

```
# Un algoritmo de derivación de palabras basado en el algoritmo de Porter, Un algoritmo para eliminar sufijos.
porter = nltk.PorterStemmer()

# Devuelve el lema más corto encontrado en WordNet.
wnl = nltk.WordNetLemmatizer()

def steam_tok(doc):

  # Array temporal para guardar tokens mientras se limpian.
  clean = []

  for tokens in doc:
    # Elimina los afijos del token y devuelve la raíz.
    clean.append(porter.stem(tokens))
    # Usamos la opcion r para lematizar adverbios.
    clean.append(wnl.lemmatize(tokens, 'r'))

  return set(clean)
```

Aplicamos las funciones anteriores

```

# ******* Inlcuye a continuación todas las líneas de código y celdas que requieras: ***********

# Se aplica función de limpieza y tokenización a los comentarios en X:
Xclean = [clean_tok(x) for x in X]

# Se aplica función de lematización a cada comentario X que previamente se limpió y tokenizó.
Xclean = [steam_tok(x) for x in Xclean]

# Imprimos los primeros cinco comentarios.
Xclean[0:5]

# *********** Aquí termina la sección de agregar código *************
```

## Diccionario

```
# a.	Usa el conjunto de entrenamiento para generar tu vocabulario
#     con un tamaño que consideres adecuado:


# ******* Inlcuye a continuación todas las líneas de código y celdas que requieras: ***********

# Se comienza por definir un nuevo diccionario para contar la frequencia con que se repite cada palabra de los comentarios.
dicc = Counter()

# Llenamos el diccionario con cada palabra y su conteo de frecuencia en los comentarios.
for k in range(len(Xtrain)):
  dicc.update(Xtrain[k])

# *********** Aquí termina la sección de agregar código *************
```

```
# b.	Indica el tamaño del vocabulario generado.

#print('Longitud del vocabulario generado:', len(dicc))

# ******* Inicia la sección de agregar código: ***********

# Iimprimos datos de nuestro diccionario.
print('Longitud del diccionario:', len(dicc))
print('word, frequency:', dicc.most_common(10))

# *********** Aquí termina la sección de agregar código *************
```

## Frecuencia minima de palabras en el diccionario

```
# Frecuencia mínima de aparición de los tokens del diccionario en el conjunto de entrenamiento.
# Se define frecuencia mínima de palabras que no se considerán en vocabulario.
min_freq = 1

# Se crea un nuevo diccionario para almacenar las palabras que cumplen con los criterios de frecuencia mínima.
dicc_min = Counter()

# Filtramos las palabras en función a su frecuencia. Si el valor es mayor que la frecuencia mínima se agrega al nuevo diccionario.
for k, v in dicc.items():
  if v > min_freq:
    dicc_min.update({k:v})

# Imprimimos la nueva longitud del nuevo diccionario.
print('Nueva longitud del nuevo vocabulario:', len(dicc_min))

# Imprimimos los primeros cinco elementos del nuevo diccionario.
print('word, frequency:', sorted(dicc_min.items(), key=lambda x: x[1])[:5], '\n')
```

## Frecuencia maxima de palabras en el diccionario

```
# Se define frecuencia máxima que se considerará dentro del diccionario con la intención de castigar a las palabras más comunes que hablan del contexto y no del sentimiento del comentario.
max_freq = 40

# Inicializamos un nuevo diccionario para  almacenar las palabras que cumplen con los criterios de frecuencia maxima.
dicc_max = Counter()

# Filtramos las palabras en función a su frecuencia. Si el valor es menor que la frecuencia maxima se agrega al nuevo diccionario.
for k, v in dicc_min.items():
  if v < max_freq:
    dicc_max.update({k:v})

# Imprimimos la nueva longitud del nuevo diccionario.
print('Nueva longitud del nuevo vocabulario:', len(dicc_max))

# Imprimimos los primeros cinco elementos del nuevo diccionario.
print('word, frequency:', sorted(dicc_max.items(), key=lambda x: x[1])[:5], '\n')
```
